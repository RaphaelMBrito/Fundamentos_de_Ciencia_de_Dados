```python
# Funções auxiliares

def transformar_df1(df):
    aux_tit=list(df)
    aux_df=df.drop(aux_tit[2:],axis=1)
    cont=0
    for k in aux_tit[2:]:
        aux_del=[]
        ano=k
        cont=cont+1
        for i in aux_tit[2:]:
            if i != ano:
                aux_del.append(i)
            novo_df = df.drop(aux_del,axis=1)
            novo_df=novo_df.rename(columns={ano:df[aux_tit[0]][0]})
            novo_df = novo_df.drop('Nome',axis=1)
            novo_df['ano']=int(ano)
        if cont>1:
            merge_df=pd.merge(aux_df,novo_df,how='outer')
            trans_df=pd.merge(trans_df,merge_df,how='outer')
        else:
            trans_df=df.drop(aux_tit[2:],axis=1)
    aux_1=df.drop(aux_tit[3:],axis=1)
    aux_1=aux_1.drop(aux_tit[0],axis=1)
    aux_1=aux_1.rename(columns={aux_tit[2]:df[aux_tit[0]][0]})
    aux_1['ano']=int(aux_tit[2])
    trans_df=pd.merge(aux_1, trans_df,how='outer')
    trans_df = trans_df.drop('Nome',axis=1)
            
    return trans_df
```
